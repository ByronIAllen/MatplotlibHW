```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_df = ride_df.merge(city_df, how='outer', on='city')

# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Break down DF by city type for further use
urban_df = combined_df.loc[combined_df['type'] == 'Urban']
rural_df = combined_df.loc[combined_df['type'] == 'Rural']
suburb_df = combined_df.loc[combined_df['type'] == 'Suburban']

# Obtain the x and y coordinates for each of the three city types
city_urb = urban_df.groupby('city')        
urb_rider = city_urb['ride_id'].count()        
x_urb = [value for value in urb_rider]
urb_fare = city_urb['fare'].mean()
y_urb = [value for value in urb_fare]    
urb_driver = city_urb['driver_count'].sum() 
size_urb = [value for value in urb_driver]

city_rural = rural_df.groupby('city')        
rural_rider = city_rural['ride_id'].count()        
x_rural = [value for value in rural_rider]
rural_fare = city_rural['fare'].mean()
y_rural = [value for value in rural_fare]
rural_driver = city_rural['driver_count'].sum() 
size_rural = [value for value in rural_driver]

city_sub = suburb_df.groupby('city')        
sub_rider = city_sub['ride_id'].count()        
x_sub = [value for value in sub_rider]
sub_fare = city_sub['fare'].mean()
y_sub = [value for value in sub_fare]
sub_driver = city_sub['driver_count'].sum() 
size_sub = [value for value in sub_driver]

# Build the scatter plots for each city types
plt.scatter(x_urb, y_urb, marker='o', facecolors='r', edgecolors='black',
           s=size_urb, alpha=0.75, label='Urban')
plt.scatter(x_rural, y_rural, marker='o', facecolors='y', edgecolors='black',
           s=size_rural, alpha=0.75, label='Rural')
plt.scatter(x_sub, y_sub, marker='o', facecolors='b', edgecolors='black',
           s=size_sub, alpha=0.75, label='Suburban')

# Incorporate the other graph properties
plt.grid()
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fares ($)')
# Create a legend
plt.legend(title='City Types', markerscale=0.3)

# Incorporate a text label regarding circle size
plt.text(42,35,'Note:\nCircle size correlates with driver count per city')

# Save Figure
plt.savefig('Images/pyber_scatter.png')
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_urb = urban_df['fare'].sum()
total_rural = rural_df['fare'].sum()
total_sub = suburb_df['fare'].sum()
totals = [total_urb, total_rural, total_sub]

# Build Pie Chart
explode = (0.1, 0, 0)
labels = ["Urban", "Rural", "Suburban"]

plt.pie(totals, explode=explode, labels=labels,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Fares by City Type')
plt.axis('equal')

# Save Figure
plt.savefig('Images/fares_pie.png')
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_urb = len(urban_df['fare'])
rides_rural = len(rural_df['fare'])
rides_sub = len(suburb_df['fare'])
rides = [rides_urb, rides_rural, rides_sub]

# Build Pie Chart
plt.pie(rides, explode=explode, labels=labels,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Rides by City Type')
plt.axis('equal')

# Save Figure
plt.savefig('Images/rides_pie.png')
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_urb = urban_df['driver_count'].sum()
driver_rural = rural_df['driver_count'].sum()
driver_sub = suburb_df['driver_count'].sum()
drivers = [driver_urb, driver_rural, driver_sub]

# Build Pie Charts
plt.pie(drivers, explode=explode, labels=labels,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Drivers by City Type')
plt.axis('equal')

# Save Figure
plt.savefig('Images/drivers_pie.png')
```

```python
# Show Figure
plt.show()
```
